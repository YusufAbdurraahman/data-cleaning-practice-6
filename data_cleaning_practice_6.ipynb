# **Data Cleaning**

```python
df = df.drop_duplicates()
print(f"Jumlah data setelah diduplikat", len(df))
```

```python
category_feature = df.select_dtypes(include = ["object"]).columns
numeric_feature = df.select_dtypes(include = ["number"]).columns
```

```python
print(category_feature)
print(numeric_feature)
```

```python
for feature in numeric_feature:
  sns.boxplot(y = df[feature])
```

```python
df.head(3)
```

```python
df.columns  = df.columns.str.replace("_"," ").str.title().str.replace(" ","_")
```
